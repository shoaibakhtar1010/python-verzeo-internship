```python
import pandas as pd
df=pd.read_csv('tmdb-movies.csv')
df.dropna
small_df=df[['budget','original_title']]
result=small_df[(small_df['budget']==df.budget.nlargest(3).iloc[-1])]
result2=small_df[(small_df['budget']==df.budget.nsmallest(3).iloc[-1])].head(1)
print('movies with third highest budget is: \n' ,result)
print ('movies with third lowest budget is:\n',result2)
```

```python
import pandas as pd
df=pd.read_csv('tmdb-movies.csv')
small_df=df[['revenue','original_title']]
result=small_df[(small_df['revenue']==df.revenue.nlargest(1).iloc[-1])]
result2=small_df[(small_df['revenue']==df.revenue.nsmallest(1).iloc[-1])].head(1)
print('movies with  highest revenue is: \n' ,result)
print ('movies with  lowest revenue is:\n',result2)
```

```python
import pandas as pd
df=pd.read_csv('tmdb-movies.csv')
df['release_date'] = pd.to_datetime(df['release_date'])
start_date = '01-01-2006'
end_date = '31-12-2006'
mask = (df['release_date'] > start_date) & (df['release_date'] <= end_date)
df = df.loc[mask]
df['runtime'].mean()
```
