common Genre for vin Diesel movie


```python
df1=df.dropna()
df1
```

```python
df2=df1[df1['cast'].str.contains('Vin Diesel')]
```

```python
str2=''
for i in df2['genres']:
    str2+=i
```

```python
k=str2.split('|')
genres_and_count={}
count=0
for j in k:
    try:
        count = genres_and_count[j]
        genres_and_count[j] = count + 1
    except:
        genres_and_count[j] = 1
        
```

```python
max(genres_and_count,key=genres_and_count.get)
```

```python
genres_and_count
```

```python
df4=df1[df1['cast'].str.contains('Emma Watson')]
str4=''
for i in df4['genres']:
    str4+=i
k=str4.split('|')
genres_and_count={}
count=0
for j in k:
    try:
        count = genres_and_count[j]
        genres_and_count[j] = count + 1
    except:
        genres_and_count[j] = 1

max(genres_and_count,key=genres_and_count.get)
```
