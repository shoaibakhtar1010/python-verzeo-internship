```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
main_df[' workclass'] = le.fit_transform(main_df[' workclass'])
main_df[' occupation'] = le.fit_transform(main_df[' occupation'])
main_df[' native-country'] = le.fit_transform(main_df[' native-country'])
main_df[' income'] = le.fit_transform(main_df[' income'])
```

```python
main_df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohc = OneHotEncoder()
```

```python
temp_df = pd.DataFrame(ohc.fit_transform(main_df[[' race']]).toarray())
```

```python
temp_df.tail()
```

```python
main_df = main_df.reset_index()
main_df
```

```python
cols = list(main_df.columns)
cols.remove('index')
main_df = main_df[cols]
main_df
```

```python
main_df2 = main_df.join(temp_df)
```

```python
main_df2.head(60)
```

```python
main_df2.columns
```

```python
X = main_df2[['age',      ' workclass',  ' education-num', ' occupation',' hours-per-week', ' native-country',' income', 0, 1, 2, 3,  4]].values
Y = main_df2[[' sex']].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, Y_train, Y_test =  train_test_split(X, Y)
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, Y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
accuracy_score(y_pred, Y_test)
```
